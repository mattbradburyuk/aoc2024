```python
# Forward only Linked List

class Link:
    def __init__(self, num, next): 
        self.num = num
        self.next = next

class LinkedList: 
    def __init__(self):
        self.len = 0
        self.first_link = None
        self.current_link = None

    def insert(self, num):
        self.len += 1
        # no links yet
        if self.first_link == None:
            link = Link(num, None)
            self.first_link = link
            self.current_link = link
        # current is last link
        elif self.current_link.next == None:
            link = Link(num, None)
            self.current_link.next = link
            self.current_link = link
        # insert into other position 
        else:
            current_next = self.current_link.next
            link = Link(num, current_next)
            self.current_link.next = link
            self.current_link = link
        return self
    
    # reset list to start and return first link
    def start(self):
        self.current_link = self.first_link
        return self.current_link
    
    # return current link
    def current(self): 
        return self.current_link

    # return the next link and move current to next
    def next(self):
        if self.current_link.next == None: return None
        else: 
            self.current_link = self.current_link.next
            return self.current_link
        
    def skip(self):
        self.current_link = self.current_link.next


    def print_list(self):
        print(f"list len: {self.len}")
        print(self.start().num, end=" ")
        while self.current_link.next != None:
            print(self.next().num, end=" ")
        print("\n")


# test LinkedList        
myll = LinkedList()
myll.insert(1).insert(30).insert(2).insert(5).insert(10)
myll.print_list()
```

```python
# fill linked List


stones = LinkedList()
for n in num_list:
    stones.insert(n)

stones.print_list()
```

```python
import math, time

def num_digits(num): return math.floor(math.log10(num)) + 1 

def has_even_digits(num): return True if num_digits(num) %2 == 0  else False

def split_num(num):

    lhs_digits = math.floor(num / 10 ** (num_digits(num)/2))
    rhs_digits = int(num - lhs_digits * 10 ** (num_digits(num)/2))
    return lhs_digits, rhs_digits

num_iter = 10

stones.print_list()

for i in range(num_iter):
    tic = time.time()
    stone = stones.start()
    cont = True
    while cont:
        if stone.num == 0: 
            stone.num += 1
        elif has_even_digits(stone.num): 
            lhs, rhs = split_num(stone.num)
            stone.num = lhs
            stones.insert(rhs)
        else:
            stone.num *= 2024

        stone = stones.next()
        if stone == None:
            cont = False


    toc = time.time()
    # print(f"Iter: {i}")
    
    print(f"iter: {i}, time take: {toc - tic}")
    stones.print_list()

print(stones.len)


```
