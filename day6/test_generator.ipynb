```python
import random


for test_no in range (10):
    rows = 10
    cols = 10
    grid = [['.' for _ in range(cols)] for _ in range(rows)]
    for i in range(10):
        r = random.randint(0,9)
        c = random.randint(0,9)
        print(r, ", ",c)
        grid[r][c] = '#'

    r = random.randint(4,6)
    c = random.randint(4,6)

    grid[r][c] = '^'

    
    with open(f"./test_inputs/test_{test_no}.txt", "w") as file:
        for row in grid:
                file.write("".join(map(str, row)) + "\n")
```
