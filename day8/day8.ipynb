```python
path = "./day8_input.txt"
# path = "./day8_test.txt"

with open(path) as file: 
    rows = file.read().splitlines()

for r in rows:
    print(r)
```

```python
num_rows = len(rows)
num_cols = len(rows[0])

# populate dictionary of antenna positions
antennas = {}
for r_ind, row in enumerate(rows):
    for c_ind, node in enumerate(row):
        if node != '.':
            if node not in antennas:
                antennas[node] = [] 
            antennas[node].append((r_ind, c_ind))
```

```python


def calc_node(first, second): 
    new_node = (2*second[0]-first[0], 2*second[1]-first[1])
    return new_node  

def checkpairs(ant_list):
    nodes = []
    for first in ant_list:
        for second in ant_list:
            if first != second: 
                node = calc_node(first, second)
                if (node[0] >= 0 and
                    node[0] < num_rows and
                    node[1] >= 0 and
                    node[1] < num_cols):
                    nodes.append(node)
    return nodes

nodes_set = set([])
for ant in antennas:
    coords = antennas[ant]
    nodes_set.update(checkpairs(coords))

print(len(nodes_set))
# for n in nodes_set:
#     print(n)
```
