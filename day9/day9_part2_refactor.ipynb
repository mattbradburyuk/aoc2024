```python
path = "./day9_input.txt"
# path = "./day9_test.txt"

with open(path) as file:
    disk_map_str = file.read()

disk_map = [int(s) for s in disk_map_str]

print(disk_map)
```

```python
data_ind = 0
block_counter = 0 
data =[]
is_data = True

for num in disk_map:
    if is_data: 
        for files_ind in range(num):
            data.append(block_counter)
        block_counter += 1
    else:
        for files_ind in range(num):
            data.append('.')

    # flip toggle
    if is_data: 
        is_data = False
    else:
        is_data = True

# target for test: 00...111...2...333.44.5555.6666.777.888899
print(data)
```

```python
# Get list of files and file indexes
files =[]
file = [data[0]]
start_ind = 0
for files_ind in range(1,len(data)):
    if data[files_ind] != data[files_ind-1]:
        files.append(file)
        file = [data[files_ind]]
        start_ind = files_ind
    else:
        file.append(data[files_ind])
files.append(file)
print(files)
```

```python
# need to stop when reach number
def find_next_space(number_file):
    file_length = len(number_file)
    for i, f in enumerate(files):
        if f[0] == number_file[0]:
            return -1
        if f[0] == '.':
            if len(f) >= file_length:
                return i
    return -1    

print("starting files: ",files, "\n")
backward_ind = len(files)-1
considered = []

while backward_ind > 0:

    # find next file with numbers in it
    if files[backward_ind][0] == '.':
        backward_ind -= 1
        continue
    # check it hasn't been considered before
    number_file = files[backward_ind] 
    if number_file[0] in considered:
        backward_ind -= 1
        continue
    considered.append(number_file[0])
    # print("number_file: ", number_file)
    
    # find the next space that's big enough
    space_ind = find_next_space(number_file)
    # print("space_ind: ", space_ind) 
    if space_ind == -1:
        backward_ind -= 1
        # print("\n")
        continue
    
    # calcaulate the filler required
    filler_length = len(files[space_ind]) - len(number_file)

    # insert the numbers file
    files[space_ind] = number_file

    # insert filler if required
    if filler_length >0:
        filler = ['.'] * filler_length
        files.insert(space_ind + 1, filler)
        backward_ind +=1 # because added an extra entry in list       
        
    # fil in '.'s
    files[backward_ind] = ['.'] * len(number_file)
    # also need to join up '.'s before and after where taken from 

    # join up spaces
    i = 0
    while i < len(files)-1:
        if files[i][0] == '.' and files[i+1][0] == '.':
            total_len = len(files[i]) + len(files[i+1])
            del files[i:i+2]
            files.insert(i, ['.'] * total_len)
            if i < backward_ind: 
                backward_ind -= 1
        else:
            i += 1

    # end of file is missing
    backward_ind -= 1
    # print("files: ", files, "\n")


# print(files[backward_ind])

print("ending files: ",files, "\n")
flatened = [f for file in files for f in file]

print(flatened)
```
