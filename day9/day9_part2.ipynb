```python
for files_ind in range(len(files)-1, 0 ,-1):
    file = files[files_ind]
    # find next space in files
    forward_ind = 0
    search = True
    while search and forward_ind + len(file[1]) < len(data) and forward_ind < file[0]:        
        # check if space at forward_ind
        space = True
        for file_ind in range(len(file[1])):
            if data[forward_ind + file_ind] != '.':
                space = False
    
        # if there is space swap the file and move the indicator forwards    
        if space:
            for file_ind in range(len(file[1])):
                # delete from end
                data[file[0] + file_ind] = '.'
                # add to space
                data[forward_ind + file_ind] = file[1][0]
            search = False 
                 
        forward_ind += 1

    print(f"files done: {files[files_ind][1][0]}") 


```
